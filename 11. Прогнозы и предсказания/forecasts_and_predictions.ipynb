## Шаг 3. Постройте модель прогнозирования оттока пользователей<a name="3"></a>  
<a class="btn btn-primary btn-sm" href="#content" role="button">Вернуться к содержанию</a>

```python
# разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = df.drop('churn', axis=1)
y = df['churn']

# разделяем модель на обучающую и валидационную выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=0, stratify=y)

# создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler()
# обучаем scaler и одновременно трансформируем матрицу для обучающей выборки
X_train_st = scaler.fit_transform(X_train)

# применяем стандартизацию к матрице признаков для тестовой выборки
X_test_st = scaler.transform(X_test)
```

#### Обучим модель с помощью логистической регрессии

```python
# выбор алгоритма
lr_model = LogisticRegression(solver='liblinear')

# обучаем модель
lr_model.fit(X_train_st, y_train)

# делаем прогнозы
predictions = lr_model.predict(X_test_st)
probabilities = lr_model.predict_proba(X_test_st)[:,1]
```

```python
print('Accuracy: {:.2f},\nPrecision: {:.2f},\nRecall: {:.2f},\nF1: {:.2f}'
      .format(accuracy_score(y_test, predictions),
              precision_score(y_test, predictions),
              recall_score(y_test, predictions),
              f1_score(y_test, predictions)))
```

#### Подберем наилучшие параметры для случайного леса

```python
# параметры для перебора
parameter_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [10, 20, 100],
    'n_estimators': [10, 20, 100, 1000]
}

rf_model = RandomForestClassifier()
grid_searcher = GridSearchCV(rf_model, parameter_grid, verbose=2)
grid_searcher.fit(X_train_st, y_train)
rf_model_best = grid_searcher.best_estimator_

print('Best params = ', rf_model_best.get_params())
```

#### Обучим модель с помощью случайного леса

```python
# выбор алгоритма с наилучшими параметрами
rf_model = RandomForestClassifier(n_estimators=1000, max_depth=20, random_state=0)

# обучаем модель
rf_model.fit(X_train_st, y_train)

# делаем прогнозы
rf_predictions = rf_model.predict(X_test_st) 
rf_probabilities = rf_model.predict_proba(X_test_st)[:,1]
```

```python
print('Accuracy: {:.2f},\nPrecision: {:.2f},\nRecall: {:.2f},\nF1: {:.2f}'
      .format(accuracy_score(y_test, rf_predictions),
              precision_score(y_test, rf_predictions),
              recall_score(y_test, rf_predictions),
              f1_score(y_test, rf_predictions)))
```

Даже полсле перебора параметров логистическая регрессия показывает себя чуть лучше, но разница практически незначительная.

## Шаг 4. Сделайте кластеризацию пользователей<a name="4"></a>  
<a class="btn btn-primary btn-sm" href="#content" role="button">Вернуться к содержанию</a>

Построим матрицу расстояний на стандартизованной матрице признаков и нарисуем дендрограмму:

```python
# создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler()
# обучаем scaler и одновременно трансформируем матрицу для обучающей выборки
X_sc = scaler.fit_transform(X)

linked = linkage(X_sc, method = 'ward') 
```

```python
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Дендрограмма')
plt.show() 
```

Судя по графику кажется, что не хватает еще одной группы. Предположим, что необходимо 5 кластеров.

Обучим модель кластеризации на основании алгоритма K-Means:

```python
# задаём модель k_means с числом кластеров 5
km = KMeans(n_clusters=5, random_state=0)

# применяем алгоритм к данным и формируем вектор кластеров 
labels = km.fit_predict(X_sc)

df['cluster_km'] = labels
```

```python
df['cluster_km'].value_counts()
```

Кластеры имеют разное количество клиентов.

```python
df.groupby('cluster_km').mean().round(2).T
```

В среднем, вероятность попасть в **отток ниже** в кластерах **0 и 4**.    
**Наибольшая вероятность оттока** в кластерах номер **3** и номер **2**.


Признаки "хорошего" клиента:
- Чем больше возраст, тем клиент постоянней;
- Чаще ходит на групповые занятия;
- Выделяется длительностью абонемента;
- Тратит на дополнительные услуги больше остальных;
- Давно ходит с момента первого посещения;
- Средняя частота посещений в неделю за все время и в предыдущий месяц больше.

Построим распределения признаков для кластеров:

```python
for column in ['gender', 'near_location', 'partner', 'promo_friends', 'contract_period']:
    plt.figure(figsize=(5, 5))
    sns.countplot(x = df[column], hue='cluster_km', data=df)
    plt.title(column)
    plt.show()
```

```python
for column in ['month_to_end_contract', 'age', 'avg_additional_charges_total', 'lifetime', 'avg_class_frequency_total', 'avg_class_frequency_current_month']:
    plt.figure(figsize=(5, 5))
    sns.boxplot(x = 'cluster_km', y = df[column], data = df)
    plt.show()
```

Те, кто уходит в отток в среднем меньше приводят своих знакомых в фитнес-центр. Их длительность абонемента чаще равна 1 месяцу. Абонемент заканчивается в ближайший месяц. Возраст ниже. Тратят на доп. услуги меньше "хороших клиентов". Меньше ходят с момента первого посещения. Меньшее количество раз ходят в фитнес в неделю.

Для каждого полученного кластера посчитаем долю оттока:

```python
round(df.groupby('cluster_km').size() / len(df), 2)
```

Самый надежный кластер - номер 1. Кластер номер 3 судя по доле является склонным к оттоку.
